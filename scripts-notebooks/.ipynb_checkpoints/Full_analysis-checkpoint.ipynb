```r
accuracy <- function(model, plotit = FALSE) {
  # Try to extract observed and predicted values
  if (inherits(model, "lm")) {
    actual <- model$model[[1]]
    predicted <- predict(model)
  } else if (inherits(model, "nls")) {
    actual <- model$m$y()
    predicted <- predict(model)
  } else {
    stop("Model type not supported. Only 'lm' and 'nls' are supported.")
  }
  
  # Check if actual and predicted are same length
  if (length(actual) != length(predicted)) {
    stop("'actual' and 'predicted' lengths differ")
  }
  
  # Calculate metrics
  residuals <- actual - predicted
  RMSE <- sqrt(mean(residuals^2))
  MAE <- mean(abs(residuals))
  # For nls, use pseudo R² (1 - SSR/SST)
  R2 <- 1 - sum(residuals^2) / sum((actual - mean(actual))^2)
  
  if (plotit) {
    plot(actual, predicted,
         xlab = "Observed",
         ylab = "Predicted",
         main = "Observed vs Predicted",
         pch = 16, col = "darkgreen")
    abline(a = 0, b = 1, col = "red", lwd = 2, lty = 2)
    legend("topleft", legend = c(
      paste("RMSE =", round(RMSE, 2)),
      paste("MAE =", round(MAE, 2)),
      paste("Eff R² =", round(R2, 3))
    ), bty = "n")
  }
  
  return(list(RMSE = RMSE, MAE = MAE, R2 = R2))
}
```

```r
# Model 30cm organic carbon stocks from shorter cores (15cm)

# The below code aims to model 30cm carbon stocks from 15cm stocks.
# We only use the subset of cores that have 30cm data to establish the relationship of 
# 30cm stocks to 15cm stocks, so that we can compare modeled to measured and derive R2, RMSE, etc.
# To find the appropriate model, linear and 2nd order polynomials are compared.
# The model summaries below allow us to determine which model fits the data best.
# We then use the best model and apply it to data from shorter cores to estimate 30cm stocks.

# Model 30cm stocks from 15cm data linear
mod15lm <- lm(data = data30nout, Corg.stock.30cm ~ Corg.stock.15cm)
summary(mod15lm) # Adj R2 = 0.9265, p < 2.2e-16
accuracy(mod15lm, plotit = TRUE) # RMSE=10, MAE=5.51
abline(a = 0, b = 1, col = "steelblue", lwd = 2)
plot(x=data30nout$Corg.stock.15cm, y=data30nout$Corg.stock.30cm, xlab = "30cm Corg stock  (Mg ha-1)", ylab = "15cm Corg stock  (Mg ha-1)", main = "Predictor: 15cm stocks, Linear model")
abline(mod15lm, col="steelblue")
```

```r
#my update
ggplot(data30nout, aes(x = Corg.stock.15cm, y = Corg.stock.30cm)) +
  geom_point(color = "black", size = 3, alpha = 0.7) +
  
  # Regression line from model (equivalent to abline(mod15lm))
  geom_smooth(method = "lm", se = FALSE, color = "steelblue", size = 1.2) +

  # 1:1 line (equivalent to abline(a = 0, b = 1))
  geom_abline(intercept = 0, slope = 1, color = "steelblue", linetype = "dashed") +

  labs(
    title = "Predictor: 15cm stocks, Linear model",
    x = "15cm Corg stock (Mg ha-1)",
    y = "30cm Corg stock (Mg ha-1)"
  ) +
  theme_minimal()
```

```r
# Model 30cm stocks from 15cm data second order
mod15nls <- nls(Corg.stock.30cm ~ a*Corg.stock.15cm+b*Corg.stock.15cm^2 , data = data30nout, start = list(a = 1, b = 1))
# Get coefficients
mod15.co <- coef(mod15nls)
# Fit function
mod15.fit <- function(x, a, b) {a*x+b*x^2}
# Plot result
plot(x=data30nout$Corg.stock.15cm, y=data30nout$Corg.stock.30cm, xlab = "30cm Corg stock  (Mg ha-1)", ylab = "15cm Corg stock  (Mg ha-1)", main = "Predictor: 15cm stocks, 2nd order poly")
curve(mod15.fit(x, a=mod15.co["a"], b=mod15.co["b"]),
      add=TRUE ,lwd=2, col="steelblue")
# Summarize model
summary(mod15nls)
accuracy(mod15nls, plotit = TRUE) # Eff R2 = 0.927, RMSE=10, MAE=4.42
abline(a = 0, b = 1, col = "steelblue", lwd = 2)

## Model comparisons
AIC(mod15lm, mod15nls) # (lm=4921, nls=4920)

# Models have similar RMSE, MAE, R2, and AIC.
```

```r
data_pred <- data %>% 
  mutate(Pred.30cm.Corg = case_when(!is.na(Corg.stock.30cm)  ~ Corg.stock.30cm,
                                    !is.na(Corg.stock.15cm)  ~ 1.9323329*Corg.stock.15cm-0.0010214*Corg.stock.15cm^2,
                                    Sampling.depth > 4 & is.na(Corg.stock.15cm) ~ (Corg.stock*(15/Sampling.depth)*1.9323329-0.0010214*(Corg.stock*(15/Sampling.depth)^2)) ),
         Pred.1m.Corg = case_when(!is.na(Corg.stock.100cm)  ~ Corg.stock.100cm,
                                  !is.na(Corg.stock.50cm)  ~ 1.91*Corg.stock.50cm-0.000499*Corg.stock.50cm^2,
                                  !is.na(Corg.stock.30cm)  ~ 3.127*Corg.stock.30cm-0.00149*Corg.stock.30cm^2,
                                    !is.na(Corg.stock.15cm)  ~ 6.058*Corg.stock.15cm-0.00521*Corg.stock.15cm^2)
         )
```
